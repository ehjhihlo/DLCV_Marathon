```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('lena.jpg')
img = img[:, :, [2, 1, 0]]
```

### 第一大題 (OpenCV 標註訓練)

1. 將 Lena 的臉蛋用紅框框起來
2. 畫出紅匡中的對角線
3. 標注紅匡左上角的座標
4. 利用 subplot 將三張圖片繪製出來

```python
# 獲得用紅匡標注的圖片
""" Your Code Here: """
img_rec = img.copy()
cv2.rectangle(img_rec, (60,40), (420,510), (0, 0, 255), 3)
# 獲得用對角線標注的圖片
""" Your Code Here: """
img_line = img.copy()
cv2.line(img_line, (60, 40), (420, 510), (0, 0, 255), 3)

# 獲得具有座標標注的圖片
""" Your Code Here: """
img_text = img.copy()
cv2.putText(img_text, '(60,40)', (60, 40), 2, 2, (0, 0, 255), 3)

# 利用 subplot 將三張圖片繪製出來
""" Your Code Here: """
plt.figure(figsize=(20,60))
imglist = [img_rec, img_line, img_text]
title = ['Rectangle', 'Line', 'Text']
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(imglist[i])
    plt.title(title[i])
    plt.axis('off')
plt.show()
```

### 第二大題 (繪製 Histogram)

1. 獲得 3 個 channel (RGB) 的 histogram
2. 利用 subplot 繪製出來

```python
# 繪製 Histogram
""" Your Code Here: """
plt.figure(figsize=(10, 8))
for i in range(3):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.subplot(1, 3, i+1, title = 'channel {}'.format(i))
    plt.bar(range(0, 256), hist[:, 0])
plt.show()
```
