```python
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt

# cv2 預設 color space 是 BGR，但 matplotlib，
# 顯示是以 RGB 來顯示，故需要做順序的調整

""" Your Code Here: 讀取 lena.png 並使用 matplotlib 顯示 """
img = cv2.imread('lena.jpg')
img = img[:, :, [2, 1, 0]]
```

# 進行放大的比較

```python
# 方便顯示與進行縮放
interpolation_lst = [cv2.INTER_NEAREST, cv2.INTER_LINEAR, cv2.INTER_AREA, cv2.INTER_CUBIC, cv2.INTER_LANCZOS4]
interpolation_name = ['INTER_LANCZOS4', 'INTER_LINEAR', 'INTER_AREA', 'INTER_CUBIC', 'INTER_LANCZOS4']

# 定義欲放大與縮小的尺寸
small_length, big_length = int(img.shape[0] / 4), int(4 * img.shape[0])
# 定義作圖版面大小
plt.figure(figsize = (25, 5))

# 開始進行5種插值法的放大比較
for i in range(len(interpolation_lst)):
    start = time.time() # 按下碼表計時
    """ Your Code Here: 插值演算法實作(Hint: cv2.resize) """
    img_scale = cv2.resize(img, None, fx=4, fy=4, interpolation=interpolation_lst[i])
    end = time.time() # 結束計時
    timer = end - start
    
    # 開始做圖
    """ Your Code Here: 開始做圖 """
    plt.subplot(1,5,i+1)
    plt.imshow(img_scale)
    plt.title(str(interpolation_name[i])+' end in '+str(round(timer,5))+'s')
plt.show()
```

# 進行縮小的比較

```python
plt.figure(figsize = (25, 5))

# 開始進行5種插值法的縮小比較
for i in range(len(interpolation_lst)):
    start = time.time() # 按下碼表計時
    """ Your Code Here: 插值演算法實作(Hint: cv2.resize) """
    img_scale = cv2.resize(img, None, fx=0.25, fy=0.25, interpolation=interpolation_lst[i])
    end = time.time() # 結束計時
    timer = end - start
    
    # 開始做圖
    """ Your Code Here: 開始做圖 """
    plt.subplot(1,5,i+1)
    plt.imshow(img_scale)
    plt.title(str(interpolation_name[i])+' end in '+str(round(timer,5))+'s')
plt.show()
```
